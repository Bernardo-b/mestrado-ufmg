### Importação das bibliotecas e dados

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df_pca = pd.DataFrame(data=X_pca, columns=['Componente 1', 'Componente 2'])
df_pca['target'] = y
```

A distribuição dos dados é a seguinte:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
scatter = plt.scatter(df_pca['Componente 1'], df_pca['Componente 2'], 
                     c=df_pca['target'], cmap='viridis')
plt.xlabel('Primeira Componente Principal')
plt.ylabel('Segunda Componente Principal')
plt.title('Visualização do Dataset de Câncer de Mama após PCA')
plt.colorbar(scatter, label='Diagnóstico (0=Maligno, 1=Benigno)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Definição das funções

```python
def gaussian_pdf_2d(x, mean, sigma):
    x = np.array(x)
    mean = np.array(mean)
    diff = x - mean
    k = len(mean)
    exponent = -0.5 * diff.T @ np.linalg.inv(sigma) @ diff
    denom = np.sqrt((2 * np.pi) ** k * np.linalg.det(sigma))
    return np.exp(exponent) / denom

def get_stats(df, target_value):
    df_target = df[df['target'] == target_value]
    media = df_target[['Componente 1', 'Componente 2']].mean().values
    std = df_target[['Componente 1', 'Componente 2']].std().values
    return media, std
```

### Resultados

Primeiramente são calculadas as estátisticas base para o problema:

```python
# Obtendo estatísticas para cada classe
media_t1, std_t1 = get_stats(df_pca, 1)
media_t0, std_t0 = get_stats(df_pca, 0)

# Criando matrizes de covariância (assumindo independência entre componentes)
sigma_t1 = np.diag(std_t1**2)
sigma_t0 = np.diag(std_t0**2)

# Testando para um ponto específico do dataset
ponto_teste = df_pca.iloc[0][['Componente 1', 'Componente 2']].values

# Calculando probabilidades
prob_t1 = gaussian_pdf_2d(ponto_teste, media_t1, sigma_t1)
prob_t0 = gaussian_pdf_2d(ponto_teste, media_t0, sigma_t0)

# Classificação bayesiana (assumindo probabilidades a priori iguais)
classe_predita = 1 if prob_t1 > prob_t0 else 0
# Obtendo estatísticas
media_t1, std_t1 = get_stats(df_pca, 1)
media_t0, std_t0 = get_stats(df_pca, 0)
sigma_t1 = np.diag(std_t1**2)
sigma_t0 = np.diag(std_t0**2)

# Calculando probabilidades para todos os pontos
df_pca['prob_1'] = df_pca.apply(lambda row: gaussian_pdf_2d(
    row[['Componente 1', 'Componente 2']].values, media_t1, sigma_t1), axis=1)
df_pca['prob_0'] = df_pca.apply(lambda row: gaussian_pdf_2d(
    row[['Componente 1', 'Componente 2']].values, media_t0, sigma_t0), axis=1)
```

Esses são os espaços de probabilidades e o espaço de decisão para o classificador com os valores de média e desvio padrão da distribuição:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Primeiro gráfico (Probabilidades)
scatter1 = ax1.scatter(df_pca['prob_0'], df_pca['prob_1'], 
                      c=df_pca['target'], cmap='viridis')
max_prob = max(df_pca['prob_0'].max(), df_pca['prob_1'].max())
ax1.plot([0, max_prob], [0, max_prob], 'r--', label='Fronteira de Decisão (p1=p2)')
ax1.set_xlabel('Probabilidade Classe 0 (Maligno)')
ax1.set_ylabel('Probabilidade Classe 1 (Benigno)')
ax1.set_title('Distribuição das Probabilidades por Classe')
plt.colorbar(scatter1, ax=ax1, label='Classe Real (0=Maligno, 1=Benigno)')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Segundo gráfico (Espaço de Decisão)
x_min, x_max = df_pca['Componente 1'].min() - 1, df_pca['Componente 1'].max() + 1
y_min, y_max = df_pca['Componente 2'].min() - 1, df_pca['Componente 2'].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

grid_points = np.c_[xx.ravel(), yy.ravel()]
prob1 = np.array([gaussian_pdf_2d(point, media_t1, sigma_t1) for point in grid_points])
prob0 = np.array([gaussian_pdf_2d(point, media_t0, sigma_t0) for point in grid_points])
Z = (prob1 > prob0).reshape(xx.shape)

ax2.contourf(xx, yy, Z, alpha=0.3, cmap='RdYlBu')
scatter2 = ax2.scatter(df_pca['Componente 1'], df_pca['Componente 2'], 
                      c=df_pca['target'], cmap='viridis', 
                      edgecolor='black', linewidth=1)
ax2.set_xlabel('Primeira Componente Principal')
ax2.set_ylabel('Segunda Componente Principal')
ax2.set_title('Espaço de Decisão do Classificador Bayesiano')
plt.colorbar(scatter2, ax=ax2, label='Classe Real (0=Maligno, 1=Benigno)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

#### Análise do Impacto dos Parâmetros nos Espaços de Decisão e Probabilidade

Os gráficos a seguir demonstram como cada parâmetro das distribuições influencia a classificação. Para cada parâmetro (médias e desvios padrão), são apresentadas três variações: o valor original e alterações de -1 e +1 unidades.

```python
def plot_parameter_effect(param_name, deltas=[-1, 0, 1]):
    fig = plt.figure(figsize=(24, 12))
    
    # Definindo título baseado no parâmetro
    param_labels = {
        'mean1': 'Média da Classe 1 (μ₁)',
        'mean2': 'Média da Classe 0 (μ₂)',
        'std1': 'Desvio Padrão da Classe 1 (σ₁)',
        'std2': 'Desvio Padrão da Classe 0 (σ₂)'
    }
    fig.suptitle(f'Efeito da Variação da {param_labels[param_name]}', 
                 fontsize=16, y=0.95)

    gs = fig.add_gridspec(2, 4, width_ratios=[1, 1, 1, 0.05])
    axes = [[fig.add_subplot(gs[i, j]) for j in range(3)] for i in range(2)]
    cax1 = fig.add_subplot(gs[0, 3])
    cax2 = fig.add_subplot(gs[1, 3])

    # Calculando estatísticas base
    m1, s1 = get_stats(df_pca, 1)
    m0, s0 = get_stats(df_pca, 0)

    for i, delta in enumerate(deltas):
        # Aplicando modificação no parâmetro escolhido
        curr_m1, curr_m0 = m1.copy(), m0.copy()
        curr_s1, curr_s0 = s1.copy(), s0.copy()
        
        if param_name == 'mean1':
            curr_m1 += delta
            title = f'μ₁ {delta:+}'
        elif param_name == 'mean2':
            curr_m0 += delta
            title = f'μ₂ {delta:+}'
        elif param_name == 'std1':
            curr_s1 = np.maximum(s1 + delta, 0.001)
            title = f'σ₁ {delta:+}'
        else:  # std2
            curr_s0 = np.maximum(s0 + delta, 0.001)
            title = f'σ₂ {delta:+}'
        
        if delta == 0:
            title = 'Original'

        sig1, sig0 = np.diag(curr_s1**2), np.diag(curr_s0**2)
        
        # Calculando probabilidades
        p1 = df_pca.apply(lambda r: gaussian_pdf_2d(
            r[['Componente 1', 'Componente 2']].values, curr_m1, sig1), axis=1)
        p0 = df_pca.apply(lambda r: gaussian_pdf_2d(
            r[['Componente 1', 'Componente 2']].values, curr_m0, sig0), axis=1)
        
        # Gráfico de probabilidades
        scatter1 = axes[0][i].scatter(p0, p1, c=df_pca['target'], cmap='viridis')
        max_p = max(p0.max(), p1.max())
        axes[0][i].plot([0, max_p], [0, max_p], 'r--', label='Fronteira de Decisão')
        axes[0][i].set_xlabel('P(0)')
        axes[0][i].set_ylabel('P(1)' if i == 0 else '')
        axes[0][i].set_title(f'Probabilidades\n{title}')
        axes[0][i].grid(True, alpha=0.3)
        axes[0][i].legend()
        
        # Gráfico de decisão
        x, y = np.meshgrid(np.linspace(df_pca['Componente 1'].min()-1, 
                                      df_pca['Componente 1'].max()+1, 50),
                          np.linspace(df_pca['Componente 2'].min()-1, 
                                      df_pca['Componente 2'].max()+1, 50))
        
        grid = np.c_[x.ravel(), y.ravel()]
        grid_p1 = np.array([gaussian_pdf_2d(p, curr_m1, sig1) for p in grid])
        grid_p0 = np.array([gaussian_pdf_2d(p, curr_m0, sig0) for p in grid])
        
        axes[1][i].contourf(x, y, (grid_p1 > grid_p0).reshape(x.shape), alpha=0.3, cmap='RdYlBu')
        scatter2 = axes[1][i].scatter(df_pca['Componente 1'], df_pca['Componente 2'], 
                                    c=df_pca['target'], cmap='viridis',
                                    edgecolor='black', linewidth=1)
        axes[1][i].set_xlabel('Componente 1')
        axes[1][i].set_ylabel('Componente 2' if i == 0 else '')
        axes[1][i].set_title(f'Espaço de Decisão\n{title}')
        axes[1][i].grid(True, alpha=0.3)

    plt.colorbar(scatter1, cax=cax1, label='Classe Real')
    plt.colorbar(scatter2, cax=cax2, label='Classe Real')
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.87)
    plt.show()

# Gerando plots para cada parâmetro
for param in ['mean1', 'mean2', 'std1', 'std2']:
    plot_parameter_effect(param)
```

É possivel perceber que os parametros de média modificam a posição do centro da menor região de decisão nos gráficos. Alterando as médias temos diferenças nas localizações das regiões de decisão mas elas não aumentam nem diminuem. Por outro lado os desvios padrões alteram muito o formato do espaço de decisão, embora as regiões de interface entre as duas categorias permaneçam aproximadamente no mesmo lugar (visto que quem controla isso é a média). Podemos perceber até que dependendo da alteração do desvio padrão, uma região concava pode ser tornar convexa e vice versa.

### Busca dos melhores parametros com AG

Implementei uma otimização com algoritimo genético para conseguir melhores parâmetros para a acurácia do modelo. Tentei otimização clássica porem não funcionou direito pois ficava preso no ponto inicial e não tinha resultados muito bons.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

def evaluate_accuracy_vectorized(params):
    m1, s1 = get_stats(df_pca, 1)
    m0, s0 = get_stats(df_pca, 0)
    
    curr_m1, curr_m0 = m1 + params[0], m0 + params[1]
    curr_s1, curr_s0 = np.maximum(s1 + params[2], 0.001), np.maximum(s0 + params[3], 0.001)
    
    # Pegando todos os pontos de uma vez
    X = df_pca[['Componente 1', 'Componente 2']].values
    
    # Calculando probabilidades vetorialmente
    diff1 = X - curr_m1
    diff0 = X - curr_m0
    
    sig1, sig0 = np.diag(curr_s1**2), np.diag(curr_s0**2)
    
    # Calculando probabilidades para todos os pontos de uma vez
    p1 = np.exp(-0.5 * np.sum(diff1 @ np.linalg.inv(sig1) * diff1, axis=1)) / np.sqrt(np.linalg.det(sig1))
    p0 = np.exp(-0.5 * np.sum(diff0 @ np.linalg.inv(sig0) * diff0, axis=1)) / np.sqrt(np.linalg.det(sig0))
    
    return accuracy_score(df_pca['target'], (p1 > p0).astype(int))

def binary_to_gray(binary):
    return binary ^ (binary >> 1)

def gray_to_binary(gray):
    binary = gray
    mask = binary
    while mask != 0:
        mask >>= 1
        binary ^= mask
    return binary

def float_to_gray(num, min_val=-3, max_val=3, bits=16):
    # Converte float para gray code
    range_val = max_val - min_val
    int_val = int((num - min_val) * (2**bits - 1) / range_val)
    return binary_to_gray(int_val)

def gray_to_float(gray, min_val=-3, max_val=3, bits=16):
    # Converte gray code para float
    binary = gray_to_binary(gray)
    range_val = max_val - min_val
    return min_val + (binary * range_val) / (2**bits - 1)

def create_individual():
    # Cria um indivíduo com 4 parâmetros codificados em gray
    return [float_to_gray(np.random.uniform(-3, 3)) for _ in range(4)]

def decode_individual(individual):
    # Decodifica um indivíduo de gray para float
    return [gray_to_float(gene) for gene in individual]


def fitness(individual):
    params = decode_individual(individual)
    return evaluate_accuracy_vectorized(params)

def tournament_selection(population, fitness_scores, k=2):
    # Seleciona um indivíduo usando torneio
    selected_idx = np.random.choice(len(population), k)
    tournament_fitness = [fitness_scores[i] for i in selected_idx]
    return population[selected_idx[np.argmax(tournament_fitness)]].copy()

def crossover(parent1, parent2, p_cross=0.7):
    # Decide se vai fazer crossover
    if np.random.random() < p_cross:
        # Crossover de um ponto
        point = np.random.randint(1, len(parent1))
        child1 = np.concatenate([parent1[:point], parent2[point:]])
        child2 = np.concatenate([parent2[:point], parent1[point:]])
        return child1, child2
    else:
        # Retorna cópias dos pais se não houver crossover
        return parent1.copy(), parent2.copy()


def mutate(individual, mutation_rate=0.05):
    # Mutação bit a bit
    for i in range(len(individual)):
        if np.random.random() < mutation_rate:
            individual[i] ^= 1 << np.random.randint(16)
    return individual

# Parâmetros do AG
pop_size = 200
n_generations = 50
elite_size = 1

# Inicialização
population = np.array([create_individual() for _ in range(pop_size)])
best_fitness_history = []
mean_fitness_history = []  # Nova lista para armazenar médias
best_individual = None
best_fitness = float('-inf')

# Evolução
from tqdm import tqdm

for gen in tqdm(range(n_generations), desc="Gerações"):
    # Avaliação
    fitness_scores = np.array([fitness(ind) for ind in population])
    
    # Atualiza melhor global e média
    gen_best_idx = np.argmax(fitness_scores)
    if fitness_scores[gen_best_idx] > best_fitness:
        best_fitness = fitness_scores[gen_best_idx]
        best_individual = population[gen_best_idx].copy()
    
    best_fitness_history.append(best_fitness)
    mean_fitness_history.append(np.mean(fitness_scores)) 
    
    # Nova população com elitismo
    new_population = [best_individual]
    
    # Completa nova população
    while len(new_population) < pop_size:
        parent1 = tournament_selection(population, fitness_scores)
        parent2 = tournament_selection(population, fitness_scores)
        child1, child2 = crossover(parent1, parent2)
        new_population.extend([mutate(child1), mutate(child2)])
    
    population = np.array(new_population[:pop_size])

# Plotando evolução do fitness
plt.figure(figsize=(12, 6))
plt.plot(best_fitness_history, 'b-', label='Melhor Fitness')
plt.plot(mean_fitness_history, 'r-', label='Média da População')
plt.xlabel('Geração')
plt.ylabel('Fitness (Acurácia)')
plt.title('Evolução do Fitness por Geração')
plt.legend()
plt.grid(True)
plt.show()

# Resultados finais
print("\nResultados Finais:")
print(f"Acurácia original: {-evaluate_accuracy_vectorized(np.zeros(4)):.3%}")
print(f"Melhor acurácia encontrada: {best_fitness:.3%}")
print(f"Média final da população: {mean_fitness_history[-1]:.3%}")
print(f"Melhores parâmetros: {decode_individual(best_individual)}")
```

Abaixo temos o resultado da otimização comparado com o resultado com os parametros originais.

```python
# Criando figura com 2x2 subplots
fig = plt.figure(figsize=(20, 16))
fig.suptitle('Comparação entre Parâmetros Originais e Otimizados', fontsize=16, y=0.95)

# Parâmetros originais e otimizados
params_orig = np.zeros(4)
params_opt = decode_individual(best_individual)

# Função auxiliar para criar os gráficos
def plot_comparison(params, position, title_prefix):
    # Calculando estatísticas
    m1, s1 = get_stats(df_pca, 1)
    m0, s0 = get_stats(df_pca, 0)
    
    curr_m1 = m1 + params[0]
    curr_m0 = m0 + params[1]
    curr_s1 = np.maximum(s1 + params[2], 0.001)
    curr_s0 = np.maximum(s0 + params[3], 0.001)
    
    sig1, sig0 = np.diag(curr_s1**2), np.diag(curr_s0**2)
    
    # Calculando probabilidades
    p1 = df_pca.apply(lambda r: gaussian_pdf_2d(
        r[['Componente 1', 'Componente 2']].values, curr_m1, sig1), axis=1)
    p0 = df_pca.apply(lambda r: gaussian_pdf_2d(
        r[['Componente 1', 'Componente 2']].values, curr_m0, sig0), axis=1)
    
    # Gráfico de probabilidades
    ax1 = plt.subplot(2, 2, position)
    scatter1 = ax1.scatter(p0, p1, c=df_pca['target'], cmap='viridis')
    max_p = max(p0.max(), p1.max())
    ax1.plot([0, max_p], [0, max_p], 'r--', label='Fronteira de Decisão')
    ax1.set_xlabel('P(0)')
    ax1.set_ylabel('P(1)')
    ax1.set_title(f'{title_prefix}\nEspaço de Probabilidades')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    plt.colorbar(scatter1, ax=ax1, label='Classe Real')
    
    # Gráfico de decisão
    ax2 = plt.subplot(2, 2, position + 1)
    x, y = np.meshgrid(np.linspace(df_pca['Componente 1'].min()-1, 
                                  df_pca['Componente 1'].max()+1, 50),
                      np.linspace(df_pca['Componente 2'].min()-1, 
                                  df_pca['Componente 2'].max()+1, 50))
    
    grid = np.c_[x.ravel(), y.ravel()]
    grid_p1 = np.array([gaussian_pdf_2d(p, curr_m1, sig1) for p in grid])
    grid_p0 = np.array([gaussian_pdf_2d(p, curr_m0, sig0) for p in grid])
    
    ax2.contourf(x, y, (grid_p1 > grid_p0).reshape(x.shape), alpha=0.3, cmap='RdYlBu')
    scatter2 = ax2.scatter(df_pca['Componente 1'], df_pca['Componente 2'], 
                          c=df_pca['target'], cmap='viridis',
                          edgecolor='black', linewidth=1)
    ax2.set_xlabel('Componente 1')
    ax2.set_ylabel('Componente 2')
    ax2.set_title(f'{title_prefix}\nEspaço de Decisão')
    ax2.grid(True, alpha=0.3)
    plt.colorbar(scatter2, ax=ax2, label='Classe Real')

# Plotando originais e otimizados
plot_comparison(params_orig, 1, f'Parâmetros Originais\nAcurácia: {-evaluate_accuracy_vectorized(params_orig):.3%}')
plot_comparison(params_opt, 3, f'Parâmetros Otimizados\nAcurácia: {-evaluate_accuracy_vectorized(params_opt):.3%}')

plt.tight_layout()
plt.show()
```

É clara a diferença entre os gráficos com parametros otimizados e os parametros originais (médias e std calculadas das distribuições). É possivel perceber que no espaço de probabilidades ele fica mais bem separado além das probabilidades ficarem com seus valroes absolutos menores. No espaço de decisão é possivel perceber que as distribuições estão mais bem divididas com um formato menos concavo/convexo para as áreas de decisão.
